```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib notebook
%config InlineBackend.figure_format='retina'
plt.style.use('ggplot')
```

```python
def get_sequence(n):
    arr = []
    while n > 1:
        arr.append(n)
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
    arr.append(1)
    
    return arr
```

```python
n = int(input())
arr = get_sequence(n)
df = pd.DataFrame(arr).rename(columns = {0: 'Value'})

fig, ax = plt.subplots()
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(arr), interval=100, repeat=True)
plt.show()
```
